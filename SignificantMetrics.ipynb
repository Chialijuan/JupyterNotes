### T-tests
https://www.youtube.com/watch?v=NF5_btOaCig&t=494s

$t = \frac{variance\,between\,groups}{variance\,within\,groups}$

Big t-value = different groups  
Small t-value = similar groups



- Independent Samples/between samples/unpaired: _testing means of 2 different groups_
    - E.g. Testing the avg quality of 2 different batches of beer
- Paired Samples/within samples/repeated measures/dependent samples: _testing means of one group twice._
    - E.g. Testing balance before and after drinking 
- One Sample: _testing mean of one group against a set of mean_
    - E.g. Testing if mean height of co-workers differ from average of 1.8m

**Limitations**  

**1. Results can only be applied to populations resembling sample that was tested.** 
- E.g. t-test on adults' cholestrol cannot be compared with childrens'
        
**2. Sample and population should be approx. normal distribution**
 
**3. Each group should have same no. of datapoints, comparing large and small groups gives inaccurate results**


**4. Independent datapoints required.**


**5. Data should be interval-level.**
- E.g. Each unit of measurement == any other unit
- Rank might not be a good unit


**6. Prone to type 1 error**
- E.g. Null hypothesis is True, but rejected

**Overcoming Limitations**

**1. Use Monte-Carlo simulations**  
Choose to simulate two groups, with five values per group, sampled from populations with means of 25 and 35 distributed according to a Gaussian distribution with a SD of 10.

Calculate the t-test and p-value of each simulation

Do ~1000 simulations and plot the P-values of each simulation on a graph


From the set of simulations, 27.55 of the simulations has p-value<0.05 == power == 27.5% < 80%. This means that there is no statistical significance.

_Power: probability of rejecting the null hypothesis when, in fact, it is false. Avoids Type 2 error of failing to reject null hypothesis_

### ANOVA

**Analysis of Variance (ANOVA) is a statistical method used to test differences between two or more means.**

- **One-way ANOVA**: only tells that at least 2 groups are statistically different from each other. Does not tell which 2 groups 

- **Post-hoc test**: to determine which specific groups differed from each other

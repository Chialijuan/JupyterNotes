### Support Vector Machine

**Supervised learning algorithm that represents points in space and separates the classes by a clear gap that is as wide as possible.**

Basically, find the optimal hyperplane which maximizes the margin from the closest points = maximize the minimum distance of the closest points to the margin.

_Usually for linear classification but can work on non-linear via **kernel trick**, which implicitly maps the inputs into high-dimensional feature space_

**Support Vectors**

Vector: Object that has both magnitude and direction

Co-ordinates of individual observations/data points 

All points $x$ on the decision boundary satisfy the
equation $w_T x + b = 0$

**Hyperplane**

- One dimension, an hyperplane is called a **point**
- Two dimensions, it is a **line**
- Three dimensions, it is a **plane**
- More dimensions you can call it an **hyperplane**

The equation of the hyperplane is defined as 
$w^Tx = 0$, which can be derived from the equation of a line: 
$y = ax + b$

Given 2 vectors $\mathbf{w}=\begin{pmatrix}-b\\-a\\1\end{pmatrix}$ and $\mathbf{x}=\begin{pmatrix}1\\x\\y\end{pmatrix}$,

$$\mathbf{w}^T\mathbf{x} = -b\times (1) + (-a)\times x + 1 \times y$$
$$\mathbf{w}^T\mathbf{x} = y - ax - b$$

We use the vector notation because: 
- easier to work with in higher dimensions
- vector $\mathbf{w}$ will always be normal to the hyperplane

**Functional margins**

Given training example $(x^{(i)}, y^{(i)})$,

$$\hat{\gamma}^{(i)} = y^{(i)}(w^Tx + b)$$

**Geometric margins**

http://cs229.stanford.edu/notes/cs229-notes3.pdf


Since $A$ represents $x^{(i)}$, we therefore find that the point B is given by $$x^{(i)}-\gamma^{(i)}\cdot\frac{w}{||w||}$$


But this point lies on the decision boundary, and all points x on the decision boundary satisfy the equation $w_T x + b = 0$. 

Hence, re-arranging to solve for $\gamma^{(i)}$ :

$$\hat{\gamma}^{(i)} = \frac{w^Tx^{(i)} + b}{||w||} = \left(\frac{w}{||w||}\right)^Tx^{(i)} + \frac{b}{||w||}$$

Given training example $(x^{(i)}, y^{(i)})$,

$$\hat{\gamma}^{(i)} = y^{(i)}\left(\left(\frac{w}{||w||}\right)^Tx^{(i)} + \frac{b}{||w||}\right)$$



Thus, group 1 is assigned a >0 value while group 2 is assigned a <0 value.

If point is very far away from the hyperplane, we are more confident of its classification

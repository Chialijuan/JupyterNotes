Metric for Linear Models: 
- $R^2$ (goes from 0 to 1) and associated p-values

$SS(fit)$ = Sum of sq for residuals to best fitting line  
$SS(mean)$ = Sum of sq for residuals to worst fitting line (mean of the y-axis values)  

$R^2 = \frac{SS(mean) - SS(fit)}{SS(mean)}$

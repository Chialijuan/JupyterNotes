### Stochastic Gradient Descent

Pros over gradient descent:

- Faster (using 1 sample per step or mini-batch)
- Able to update parameters using new data without redoing

**Gradient Descent could take a long time if there were many data points because we have to plug ALL the _Observed_ data into the Loss Function.** 

**E.g. Our model is a Logistic Regression that uses 23,000 genes and 1 million data points to predict if someone has cancer. Then we would need to calculate for 23,000 derivates and plug in 1 million terms for each derivate!**
